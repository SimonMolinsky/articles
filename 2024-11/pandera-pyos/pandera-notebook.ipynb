```python
schema = pa.DataFrameSchema(
    {
        "name": pa.Column(str, nullable=True),
        "host_id": pa.Column(str),
        "host_name": pa.Column(str, nullable=True),
        "neighbourhood_group": pa.Column(str),
        "neighbourhood": pa.Column(str),
        "latitude": pa.Column(float),
        "longitude": pa.Column(float),
        "room_type": pa.Column(str),
        "price": pa.Column(float, checks=[check_price_gt]),
        "minimum_nights": pa.Column(int),
        "number_of_reviews": pa.Column(int, nullable=True),
        "last_review": pa.Column(str, nullable=True),
        "reviews_per_month": pa.Column(float, nullable=True),
        "calculated_host_listings_count": pa.Column(int, nullable=True),
        "availability_365": pa.Column(int, nullable=True)
    },
    index=pa.Index(int),
    strict=True,
    coerce=True
)
```

```python
check_neighborhood = pa.Check.isin(['Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'Bronx'])
check_latitude_ge = pa.Check.ge(40.50)
check_latitude_le = pa.Check.le(40.91)
```

```python
schema = pa.DataFrameSchema(
    {
        "name": pa.Column(str, nullable=True),
        "host_id": pa.Column(str),
        "host_name": pa.Column(str, nullable=True),
        "neighbourhood_group": pa.Column(str, checks=[check_neighborhood]),
        "neighbourhood": pa.Column(str),
        "latitude": pa.Column(float,
                              checks=[
                                  check_latitude_ge,
                                  check_latitude_le
                              ]),
        "longitude": pa.Column(float),
        "room_type": pa.Column(str),
        "price": pa.Column(float, checks=[check_price_gt]),
        "minimum_nights": pa.Column(int),
        "number_of_reviews": pa.Column(int, nullable=True),
        "last_review": pa.Column(str, nullable=True),
        "reviews_per_month": pa.Column(float, nullable=True),
        "calculated_host_listings_count": pa.Column(int, nullable=True),
        "availability_365": pa.Column(int, nullable=True)
    },
    index=pa.Index(int),
    strict=True,
    coerce=True
)
```
